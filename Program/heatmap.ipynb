```python
# 规范化数据

#读取聚类结果.csv文件,建立dataframe
cityAverage = pd.read_csv('../聚类结果.csv',encoding='utf-8')
new_cityAverage = cityAverage.copy()
new_cityAverage['城市'] = new_cityAverage['城市'] + '市'

#利用city_province,在new_cityAverage的最后添加一列province，用于存放城市所属的省份
new_cityAverage['省份'] = new_cityAverage['城市'].map(city_province)
```

```python
provinceAverage = new_cityAverage.groupby('省份').agg({'运营线路条数':'sum', '运营里程（公里）':'sum', '客运量（万人次）':'sum'}).reset_index()
cols = ['province','line_number','operating_mileage','passenger_volume']
provinceAverage.columns = cols
```

```python
prov_data = pd.merge(prov_area, provinceAverage, on = 'province', how='outer')
prov_data['density'] = 100*prov_data['operating_mileage'] / ((prov_data['area'])**0.5)
prov_data['pressure'] = prov_data['passenger_volume'] / prov_data['operating_mileage']
prov_data = prov_data.sort_values(by='density',ascending=False)
prov_data.fillna(0,inplace=True)
prov_data
```
